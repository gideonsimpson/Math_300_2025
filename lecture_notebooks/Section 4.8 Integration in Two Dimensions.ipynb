```julia
using Trapz
using Plots
```

Note that for 3D visualization we need to apply a transpose 

```julia
x1 = LinRange(0, 1, 11)
x2 = LinRange(0, 1, 11)

f=x-> sinpi(x[1])^2 * exp(x[2]);
z = [f([x1_, x2_]) for x1_ in x1, x2_ in x2]

@show size(z);

surface(x1, x2, z', xlabel="x", ylabel="y", zlabel="z")
```

```julia
function trapz2d(f, Δx, Δy)
    m, n = size(f)
    s = 0.0
    for j in 1:m-1, i in 1:n-1
        s += 0.25 * (f[j, i] + f[j+1, i] + f[j, i+1] + f[j+1, i+1])
    end
    return s * Δx * Δy
end
```

```julia
@show trapz2d(z, step(x1), step(x2));
```

```julia
@show trapz((x1, x2), z)
```
